```python
class Simulation:
    def __init__(self, initial, R_0, beta_para):
        self.infected = initial
        self.incubation = np.random.beta(3, 11, size=initial) * 21
        self.day = 0
        self.R_0 = R_0
        self.symptom = 0
        self.symptom_history = []
        self.beta_para = beta_para
        
    def step(self):
        self.day += 1
        new_infected = int(self.infected * (self.R_0 - 1))
        self.incubation = np.append(self.incubation, np.random.beta(self.beta_para[0], self.beta_para[1], size=new_infected) * self.beta_para[2] + self.day)
        self.infected += new_infected        
        self.symptom = (self.incubation <= self.day).sum()
        self.symptom_history.append(self.symptom)
```

```python
s = Simulation(4, 1.46, (3,11,21))
limit_time = 20
end_time = 20

for _ in range(limit_time):
    s.step()
    
s.R_0 = 1.18
for _ in range(end_time):
    s.step()

plt.figure(figsize=(8,6))
plt.plot(np.diff(s.symptom_history))
plt.axvline(limit_time, c='orangered',ls='--')
```
