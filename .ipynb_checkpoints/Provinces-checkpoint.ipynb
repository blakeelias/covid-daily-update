```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import itertools
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('data/2019-nCov stats - China Provinces.csv').set_index('Unnamed: 0').iloc[:,:33].dropna()
for i in list(df):
    df[i] = df[i].astype(int)
    df['d_'+i] = df[i] - df[i].shift(1)
df
```

```python
for i in list(df)[1:]:
    focus = df[[i]].copy()
    focus['new'] = focus[i] - focus[i].shift(1)

    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
    ax.plot(focus, label=i)
    ax.set_yscale('log')
    ax.tick_params(labelrotation=90)
    ax.legend()
    plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, sharey=True, figsize=(8,8))

province = 'Guangdong'

ax.set_yscale('log')
ax.plot(df.index, df[province], label=province+' total confirmed')
ax.scatter(df.index, df[('d_'+province)], c='orangered', s=5, label=province+' daily confirmed')
ax.legend()

for a, b in [(1,11),(11,18),(17,36)]:
        slope, intercept, _,_,_ = stats.linregress(np.arange(a,b),np.log(df[('d_'+province)][a:b].values))
        x = np.arange(a,b)
        y = np.exp(slope*x + intercept)
        ax.plot(x, y)
        ax.annotate(df.index[b-1]+'\n'+(str(np.round(np.exp(slope),2))), xy=(x[-1],y[-1]), fontsize=12)
ax.tick_params(labelrotation=90, labelsize=14)
ax.xaxis.set_major_locator(plt.MaxNLocator(15))
plt.tight_layout()
```

```python
df_k = pd.read_csv('data/2019-nCov stats - SK.csv').set_index('Unnamed: 0')
for i in list(df_k):
    df_k['d_' + i] = df_k[i] - df_k[i].shift(1)
df_k.dropna(subset=['Daegu'], inplace=True)
df_k
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, sharey=True, figsize=(10,8))

for province in ['Seoul','Daegu','Gyeongbuk','Gyeongnam']:
# province = 'Gyeongnam'

    ax.set_yscale('log')
    # ax.plot(df_k.index, df_k[province], label=province)
    ax.scatter(df_k.index, df_k[('d_'+province)], s=45, label=province)
    ax.legend(prop={'size': 12})

    for a, b in [(1,4)]:
            slope, intercept, _,_,_ = stats.linregress(np.arange(a,b),np.log(df_k[('d_'+province)][a:b].values))
            x = np.arange(a,b)
            y = np.exp(slope*x + intercept)
            ax.plot(x, y)
            ax.annotate(np.round(np.exp(slope),2), xy=(1.6,np.exp(slope*1.6+intercept+0.3)), fontsize=14)
    ax.tick_params(labelrotation=90, labelsize=14)

    # ax.xaxis.set_major_locator(plt.MaxNLocator(21))
    ax.set_ylim(bottom=0.2)    
    plt.tight_layout()
```
